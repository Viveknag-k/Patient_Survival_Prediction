```python
import numpy as np
import pandas as pd
data = pd.read_csv("C:\\Users\\Vivek Nag Kanuri\\Downloads\\Dataset.csv")
data.head()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
df=data.copy()
```

```python
df.interpolate(inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(['encounter_id','patient_id','hospital_id'],axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df.columns:
    if df[i].dtypes==object:
        df[i]=l.fit_transform(df[i])       
```

```python
x=df.drop('hospital_death',axis=1)
y=df['hospital_death']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
import tensorflow as tf
import keras 
from keras.models import Sequential
from keras.layers import Dense, Activation

model = Sequential()
model.add(Dense(64, input_dim=182))
model.add(Activation('relu'))
model.add(Dense(32))
model.add(Activation('relu'))
model.add(Dense(16))
model.add(Activation('relu'))
model.add(Dense(1))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history=model.fit(X_train, y_train, epochs=25, batch_size=64)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
a=accuracy_score(pred,y_test)
print('Accuracy : ',a*1000)
```
